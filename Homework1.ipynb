```python
import matplotlib.pyplot as plt
import numpy as np

# X 값 생성
X = np.linspace(-10, 10, 100)

# 시그모이드 함수 계산
Y = 1 / (1 + np.exp(-X))

# 시그모이드 함수의 미분 계산
sigmoid_derivative = Y * (1 - Y)

# 그래프 그리기
plt.plot(X, Y, label="Sigmoid Function")
plt.plot(X, sigmoid_derivative, label="Sigmoid Derivative", linestyle="dashed")

plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.title("Sigmoid Function and Its Derivative")

plt.show()
```
